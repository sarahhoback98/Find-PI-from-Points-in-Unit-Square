```python
import random
import sys
from math import pi, sin, cos, sqrt
import matplotlib.pyplot as plt
from __future__ import division
import numpy as np
import numpy
import matplotlib.mlab as mlab
from matplotlib.patches import Circle
import time
import math

x_pts =[]
y_pts =[]
xy_pts =[]
def random_points_generator(x):
    for i in range (x):
        x_points= random.uniform(0, 1)
        y_points= random.uniform(0, 1)
        x_pts.append(x_points)
        y_pts.append(y_points)
```

```python
def iterations(x):
    random_points_generator(x) 
    pts1= np.column_stack((x_pts, y_pts))
    circ = Circle((0, 0), radius = 1)
    y = 0
    y= 0+x
    counter1=0.0
    counter = 0
    for i in range (y):
        array1 = pts1[i]
        truefalselist1 =[]
        truefalselist1= (circ.contains_point(array1))

        if (sum([truefalselist1])) > 0:
           counter +=1
    counter1= ((counter)/x)*4
    return counter1
```

```python
iterations(10000)
```

```python
difference = list()

def values_for_log(x):
    for i in range (1,x):
        iterations(10**i)
        difference_value = abs(iterations(10**i)-math.pi)
        difference.append(difference_value)
        
        

    return difference
```

```python
values_for_log(6);
xnew= list(range(1,6))
print (difference)
```

```python
plt.plot(xnew,difference)
plt.legend(['Logarithmic Scale of Pi Approximations'])
```

```python
time_for_iterations= []
def timevalues_iteration(x):
    for i in range (1,x):
        start = time.time()
        iterations(10**i)
        end = time.time()
        time_for_iterations.append((end - start))
```

```python
timevalues_iteration(6); 
x1= list(range(1,6))
```

```python
plt.plot(x1,time_for_iterations)
plt.legend(['Time for Each Iteration of Computing Pi'])
```
